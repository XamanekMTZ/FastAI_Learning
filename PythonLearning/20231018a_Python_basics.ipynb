# Números

## Enteros / Integers

Los enteros en Python tienen un rango ilimitado, estan sujetos únicamente a la cantidad de memoria virtual disponible en el equipo. Esto quiere decir que no importa que tan grande sea el número que quieras almacenar, mientras tengas suficiente memoria en tu computadora Python se encargara de él.

Los Integers / Enteros pueden ser positivos, negativos o 0 (cero), y con ellos podemos realizar las más básicas de las operaciones matemáticas como vemos abajo.

````python
# Aqui declaramos una variable con valor 14
a = 14

# Aqui declaramos una variable con valor 3
b = 3
````

````python
# Realizamos una suma o adición
a + b
````

````python
# Realizamos una resta ó sustracción
a - b
````

````python
# Realizamos una multiplicación
a * b
````

````python
# Realizamos una división verdadera
a / b
````

````python
# Realizamos una división entera
a // b
````

````python
# Realizamos una operación módulo que nos devuelve el residuo de una división
a % b
````

````python
# Elevamos a una potencia
a ** b
````

Ahora veamos que tal se comportan las divisiones cuando introducimos números negativos

````python
# División real
7 / 4
````

````python
# División entera, al truncar nos devuelve 1
7 // 4
````

````python
# División real de nuevo, pero ahora esta en negativo
-7 / 4
````

````python
# División entera, el resultado no es el mismo que el anterior pero en negativo
-7 // 4
````

Pero ¿Por qué en el último ejemplo no nos devolvio -1?

Python trabaja de esta manera, siempre redondea la división entera hacia infinito negativo

Las potencias en Python funcionan tanto con lo operandos
```base ** potencia```
como con la función
```pow()```

````python
pow(10, 3)
````

````python
10 ** 3
````

````python
pow(10, -3)
````

````python
10 ** -3
````

## Booleans
La algebra booleana es un subtema del algebra en la que los valores de las variables son ```verdadero``` o ```falso``` y también los podemos interpretar como ```1``` y ```0``` respectivamente, los booleanos son un subconjunto de los enteros, es por esto que pueden ser representados con números 1 y 0, en código Python escribimos ```True``` y ```False```  

````python
# Si queremos demostrar esto, utilizaremos la función int() para determinar el valor de  True o False
# True se comporta como un 1
int(True)
````

````python
# False se comporta como un 0
int(False)
````

````python
# 1 se puede evaluar como un True si usamos la función bool()
bool(1)
````

````python
# Así como cualquier número negativo
bool(-42)
````

````python
# Como es de esperarse, 0 (cero) se evalua a False
bool(0)
````

Así, podemos realizar operaciones de algebra booleana con los operadores AND, OR y NOT (que en otros lenguajes pueden ser representados con && para AND, || para OR y ! para NOT)

````python
# Así podemos negar un valor verdadero con
not True
````

````python
# Negamos un valor false con
not False
````

````python
# Si dos valores son verdaderos, el operador AND devuelve verdadero
True and True
````

````python
# Si uno de los valores fuese falso
True and False
````

````python
# Si usamos el operador OR, al menos uno de los valores debe ser verdadero para que devuelva verdadero
True or False
````

````python
# Si todos los valores son falsos devuelve falso
False or False
````

Entonces, si intentamos utilizar ```True``` y ```False``` en operaciones matemáticas

````python
# Es como sumar 1 + 1
1 + True
````

````python
# Aqui es como sumar 0 + 42
False + 42
````

````python
# Aqui es como restar 7 - 1
7 - True
````

# Números reales
Números reales o de punto flotante, estan representados en Python de acuerdo al formato IEEE 754 de doble precisión binaria de punto flotante, que se encuentra almacenado en 64 bits de información dividido en tres secciones


1.   Signo
2.   Exponente
3.   Mantisa (o lo que sigue a la derecha del punto decimal)

Muchos otros lenguajes de programación usan dos diferentes formatos, de precisión sencilla y doble, la primera toma 32 bits de memoria y la segunda 64 bits, Python solo utiliza la segunda forma que es de 64 bits de memoria.

Un ejemplo debajo



````python
# Declaramos una variable con punto decimal
pi = 3.1415926536
radio = 4.5
area = pi * ( radio ** 2)
area
````

Python tiene una pequeña función que nos permite ver como se comportarían los punto flotante en nuestros equipos

````python
import sys # aqui importamos la librería 'sys'
sys.float_info
````

Tenemos un total de 64 bits para representar valores de punto flotante, esto significa que podemos representar al menos 2 a la 64va potencia (es decir 18,446,744,073,709,551,616) números distintos.

# Números complejos
Python nos da soporte para trabajar con números complejos, si no los conoces aún, los números complejos son aquellos que pueden ser expresados como "```a + ib```" donde ```a``` y ```b``` son números reales y ```i``` es un número imaginario, por ejemplo la raiz cuadrada de ```-1```.

````python
# Declaramos una variable
c = 3.14 + 2.73j
c
````

````python
# Lo mismo pero declarado de otra manera
b = complex(3.14, 2.73)
b.real # la parte real
````

````python
b.imag # la parte compleja o imaginaria
````

````python
b.conjugate()
````

````python
b ** 2
````

````python
d = 1 + 1j
b - d
````

# Fracciones y decimales
Terminamos el tour de números con fracciones y decimales.
Las fracciones muestran un número racional en su numerador y denominador en la forma más baja.

````python
from fractions import Fraction # importamos la libreria
Fraction(10, 6) # al mandar a imprimir esto, veremos que la fracción (el 10 siendo el numerador y el 6 el denominador) ha sido simplificada
````

````python
Fraction(1, 3) + Fraction(2, 3) # Esto es el equivalente a 1/3 + 2/3 == 3/3 == 1/1
````

````python
f = Fraction(10, 6)
f.numerator
````

````python
f.denominator
````

````python
f.as_integer_ratio()
````

El método ```as_integer_ratio()``` nos permite trabajar tanto con enteros y booleanos sin preocuparnos que tipo de número estemos utilizando para nuestras operaciones.

Aunque las fracciones son muy útiles, no es algo común de ver en software comercial, para esos casos es mucho más común utilizar números decimales donde la precisión lo es todo por ejemplo en cálculos científicos y financieros.

Veamos un pequeño ejemplo.

````python
from decimal import Decimal as D # podemos importar una libreria y renombrarla con algo corto para no estar utilizando todo el nombre cada vez que la utilicemos
D(3.14) # Pi, como punto flotante, encontramos problemas de aproximación
````

````python
D('3.14') # Pi, como "String" (ya veremos esto en la siguiente clase), sin problemas de aproximación
````

````python
D(0.1) * D(3) - D(0.3) # Como flotante, de nuevo con problemas de aproximacion
````

````python
D('0.1') * D(3) - D('0.3') # Como "String", todo perfecto
````

````python
D('1.4').as_integer_ratio() # 7/4 = 1.4 (Transformamos un número en su fracción)
````

Es notable como al construir un decimal desde un número flotante nos encontramos con problemas de aproximación que vienen del manejo de números flotantes (inherentemente, las computadoras tienen muchos problemas al procesar el punto flotante desde siempre), en cambio si construimos el decimal desde un entero o una representación en "String" de un número no tenemos estos problemas de aproximación. Por estas razones cuando sea necesario e importante tener una precisión exacta, usaremos decimales.
